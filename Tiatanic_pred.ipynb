```python
# dropping the cabin column

dataset.drop(columns=["Cabin"], axis=1, inplace=True)
```

We drop the cabin column since there are 687 null values for a total of 891 passegers.

```python
# filling the null values of the age column of dataset since age is
# an important factor in deciding whether the passenger survived or not.
#also filling the column embarked column
dataset["Age"].fillna(dataset["Age"].mean(), inplace=True)
dataset["Embarked"].fillna(dataset["Embarked"].mode()[0], inplace=True)
```

```python
dataset.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
dataset
```

```python

X=dataset.drop(columns=['PassengerId','Name','Ticket'],axis=1)
```

```python
Y=dataset['Survived']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python

print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python

model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

```python

print(X_train_prediction)
```

```python

train_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python

print("Accuracy Score of training data: ",train_data_accuracy)
```

```python

X_test_prediction=model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python

print("Accuracy score of testing data:",test_data_accuracy)
```
